```python
from time import *
def all_permutations1(myset):
    if(len(myset) == 1):
        return [list(myset)]
    perms = []   
    for x in myset:
        newset = myset - { x }
        subperm = all_permutations1(newset)
        for y in subperm:
            p =  [x]
            p.extend(y)
            perms.append(p)
    return perms


def all_permutations2(myset):
    if(len(myset) == 1):
        return [list(myset)]
    perms = []   
    for x in myset:
        newset = myset - { x }
        subperm = all_permutations2(newset)
        for y in subperm:
            p =  [x]
            y.extend(p)
            perms.append(p)
    return perms


n=4
t0 = clock()
p1 = all_permutations1(set(range(1, n+1)))
t1 = clock()
p2 = all_permutations1(set(range(1, n+1)))
t2 = clock()
print("method 1 time cost:", t1-t0)
print("method 2 time cost:", t2-t1)
```
